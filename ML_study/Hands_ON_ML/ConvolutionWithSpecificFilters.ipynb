```python
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_sample_image
```

## Load the Sample Image

```python
china = load_sample_image("china.jpg")/255
```

```python
image_original = np.array(china)
```

```python
image = image_original[tf.newaxis, :]
```

```python
image.shape
```

```python
batch_size, height, width, channels = image.shape
```

## Create a filter

```python
filter = np.zeros(shape = (7,7,channels,2), dtype = np.float32)
```

```python
filter
```

```python
filter.shape
```

```python
filter[:,3,:,0] = 1 # vertical
```

```python
filter[3,:,:,1] = 1 # horizental
```

```python
filter
```

## Convolution

```python
output = tf.nn.conv2d(image, filter, strides = 1, padding = "SAME")
```

```python
output.shape
```

## Visualize

```python
plt.figure(figsize=(15,7))

plt.subplot(231)
plt.imshow(image_original)
plt.subplot(232)
plt.imshow(output[0,:,:,0], cmap = "gray")
plt.subplot(233)
plt.imshow(output[0,:,:,1], cmap = "gray")

plt.subplot(235)
plt.imshow(filter[:,:,0,0], cmap = "gray") # vertical filter
plt.subplot(236)
plt.imshow(filter[:,:,0,1], cmap = "gray") # horizontal filter

plt.show
print()
```
