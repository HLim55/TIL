```python
import tensorflow as tf
import keras
import numpy as np

from sklearn.datasets import load_sample_image

import matplotlib.pyplot as plt
%matplotlib inline
```

## Image Load (sklearn.datasets load_sample_image)

```python
flower_origin = load_sample_image("flower.jpg")/255
flower = np.array(flower_origin)
print(flower.shape)

plt.figure(figsize = (10,5))
plt.imshow(flower)

flower = flower[tf.newaxis,...]
print(flower.shape)

image = tf.convert_to_tensor(flower)
print(image.shape)
```

## Define Pooling (Keras)

```python
max_pool_wo_Pad = keras.layers.MaxPool2D(pool_size = (2,2), padding = 'valid')
```

```python
max_pool_w_Pad = keras.layers.MaxPool2D(pool_size = (2,2), padding = 'same')
```

```python
avg_pool_wo_Pad = keras.layers.AveragePooling2D(pool_size = (2,2), padding = 'valid')
```

```python
avg_pool_w_Pad = keras.layers.AveragePooling2D(pool_size = (2,2), padding = 'same')
```

## Pooling Results

```python
output_MaxPool_wo_Pad = max_pool_wo_Pad(image)
output_MaxPool_w_Pad = max_pool_w_Pad(image)
output_AvgPool_wo_Pad = avg_pool_wo_Pad(image)
output_AvgPool_w_Pad = avg_pool_w_Pad(image)
```

```python
print(output_MaxPool_wo_Pad.shape)
print(output_MaxPool_w_Pad.shape)
print(output_AvgPool_wo_Pad.shape)
print(output_AvgPool_w_Pad.shape)
```

## Visualize

```python
output_MaxPool_w_Pad = tf.squeeze(output_MaxPool_w_Pad, axis = 0)
```
