```python
num_classes = 10
```

```python
from tensorflow import keras
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape
```

이미지 하나를 3차원 array 형태로 변환하는 것이 필요 <br>
28x28x1 

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
x_train = x_train.astype('float32')/255 # 0 ~ 1 사이의 값으로 normalization 하기
x_test = x_test.astype('float32')/255
```

```python
from keras.utils import to_categorical 
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28,28,1))) # strides = (1,1)
```

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(128, activation='relu'))
```

```python
model.add(Dense(num_classes, activation = 'softmax'))
```

```python
model.summary()
```

```python
keras.utils.plot_model(model)
# pip install pydot
# pip install pydotplus
# pip install graphviz
```

```python
keras.utils.plot_model(model, show_shapes=True, to_file='cnn_simple.png')
```

```python
from tensorflow.keras import optimizers
model.compile(optimizer=optimizers.RMSprop(lr=0.001),
               loss='categorical_crossentropy',
               metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train_one_hot, epochs=10, batch_size=128, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','val'])
plt.show()
```

```python
model.evaluate(x_test, y_test_one_hot)
```
